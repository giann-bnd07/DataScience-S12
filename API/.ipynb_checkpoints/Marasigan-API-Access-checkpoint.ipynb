# Getting data from a news site using API

for the extraction of data from a news site, I have decided on using Inquirer.net as the disallows from the robot.txt of the website is not as strict compared to other news websites

I have also used newsAPI in the extraction of data from inquirer, as well as the mixing of web scraping, with the use of the requests package.

```python
import requests
import json
import pprint
import pandas as pd
from newsapi import NewsApiClient
```

```python
#api key: 051a9b95d44f4c3486b6e47abc8f45ca
key = "051a9b95d44f4c3486b6e47abc8f45ca"
url11 = ('https://newsapi.org/v2/everything?'
       'domains=inquirer.net&'
       'from=2021-03-11&'
       'to=2021-03-11&'
       'language=en&'
       'pageSize=50&'
       'apiKey=051a9b95d44f4c3486b6e47abc8f45ca')

response_mar_11 = requests.get(url11)
```

```python
url12 = ('https://newsapi.org/v2/everything?'
       'domains=inquirer.net&'
       'from=2021-03-12&'
       'to=2021-03-12&'
       'language=en&'
       'pageSize=50&'
       'apiKey=051a9b95d44f4c3486b6e47abc8f45ca')

response_mar_12 = requests.get(url12)
```

```python
import datetime
```

```python
Mar11 = json.loads(response_mar_11.content)
Mar11_Articles = Mar11['articles']
Mar11_json = []

Mar12 = json.loads(response_mar_12.content)
Mar12_Articles = Mar12['articles']
Mar12_json = []

start = 0
```

```python
def format_date(string):
    Da_Date = datetime.datetime.strptime(string, "%Y-%m-%dT%H:%M:%SZ")
    return Da_Date
```

```python
for i in range(start, len(Mar11_Articles), 1):
    dlist   = format_date(Mar11_Articles[i]['publishedAt'])
    temp    = dlist.date()
    date    = str(temp)
    title   = Mar11_Articles[i]['title']
    article = Mar11_Articles[i]['content']
    author  = Mar11_Articles[i]['author']
    Mar11_json.append({
            "date": date,
            "title": title,
            "article": article,
            "author": author
        })

Mar11_json
```

```python
for i in range(start, len(Mar12_Articles), 1):
    dlist   = format_date(Mar12_Articles[i]['publishedAt'])
    temp    = dlist.date()
    date    = str(temp)
    title   = Mar12_Articles[i]['title']
    article = Mar12_Articles[i]['content']
    author  = Mar12_Articles[i]['author']
    Mar12_json.append({
            "date": date,
            "title": title,
            "article": article,
            "author": author
        })

Mar12_json
```

```python
with open('Inquirernews_mar11_2021.json','w') as outfile:
    json.dump(Mar11_json, outfile)

with open('Inquirernews_mar12_2021.json','w') as outfile:
    json.dump(Mar12_json, outfile)
```
